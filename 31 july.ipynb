```python
import re
def emailValidator(email):
    pattern = '^[a-z][a-z0-9_-]{5,16}[@][a-z0-9]{2,10}[.][a-z]{1,4}$'
    if re.match(pattern, email):
        print("Valid")
    else:
        print("Invalid")
    
emailValidator('abc456@gmail.com')
emailValidator('456abc@yahoo.com')
emailValidator('abc_456@gitam.ed1')
emailValidator('abc-456@abc-d.in')
emailValidator('python@python.edu')
emailValidator('abc 456@edu.edu')
```

```python
# Transpose of Matrix...
x = [[12,7],
    [4, 5],
    [3,8]]
result = [[0, 0, 0],
         [0, 0, 0]]
for i in range(len(x)):
    for j in range(len(x[0])):
        result[j][i] = x[i][j]

for r in result:
    print(r)
```

```python
#happy Number
def happyNum(n):
    sum = rem = 0
    for i in range(2):
        while n > 0:
            rem = n % 10
            sum = sum + (rem * rem)
            n = n // 10
    return sum
while happyNum(13) != 1:
    result = happyNum(13)
    if happyNum(result) ==1:
        print("happy num")
        break
    

    

```
