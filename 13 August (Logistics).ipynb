```python
import statsmodels.formula.api as smf
logit_model = smf.logit('insuranceclaim ~ age+sex+bmi+children+smoker+region+charges', data = claim).fit()
```

```python
type(logit_model)
logit_model.params
```

```python
# summary
logit_model.summary()
```

```python
y_pred1 = logit_model.predict(claim)
y_pred1
```

```python
claim["pred_prob"] = y_pred1
claim.head()
```

```python
claim["PValue"] = 0
claim.head()
```

```python
claim.loc[claim.pred_prob >= 0.5,'PValue'] = True
claim.head()
```

```python
# confusion matrix 
confusion_matrix = pd.crosstab(claim['insuranceclaim'],claim.PValue)
confusion_matrix
```
